```python
import pandas as pd # pandas is a library and it is related to database and we can work it on excel things and pd is a short form of pandas using "as" keyword
```

```python
dataset = pd.read_csv("Salary_Data.csv") #Already imported excel in local can call using "read_csv" function
```

```python
dataset #It will show the output of excel sheet
```

```python
independent = dataset[["YearsExperience"]] #Input as indepedent
dependent = dataset[["Salary"]] #output as dependent
independent
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split #Machine learning algorithm provider is sklearn,they have "n" number of libraries (contains classes and functions)(For avoiding memory storage,We fetched particular class and function over here)
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) #we splited train and test dataset.
```

```python
X_train,X_test,Y_train,Y_test
```

```python
from sklearn.linear_model import LinearRegression 
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
weight = regressor.coef_ # coffecient of weight known as slope,constant distance
bias=regressor.intercept_ # minimun value,starting of the line
weight,bias
```

```python
Y_pred = regressor.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import r2_score #libraries of evaluating parameters
r_score = r2_score(Y_test,Y_pred) #We give actual output,predicted value.r2_score function act as a parameterized function
r_score  #result is nearly 1,so it is better performance model
```

```python
#libraries of read and write the excel
import pickle
filename="FinalizedModel_For_SalaryPrediction.sav" #should save as .sav format and need to assign variable.
pickle.dump(regressor,open(filename,"wb")) #wb is write binary and it will dump the regressor model in above file.
```

```python
loaded_model=pickle.load(open("FinalizedModel_For_SalaryPrediction.sav","rb")) #rb is read-binary 
result = loaded_model.predict([[30]])
result
```
