```python
import pandas as pd
```

```python
Dataset = pd.read_csv("50_Startups.csv")
```

```python
Dataset
```

```python
dataset=pd.get_dummies(Dataset,dtype=int,drop_first=True )
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
dependent = dataset[["Profit"]]
```

```python
import lightgbm
```

```python
!pip install lightgbm
```

```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMRegressor
params_grid={'boosting_type':['dart','gbdt'],'n_estimators':[10,100,1000,2000]}

regressor = GridSearchCV(LGBMRegressor(),params_grid,refit=True,verbose=3,n_jobs=-1)
regressor.fit(independent,dependent)
```

```python
re=regressor.cv_results_
```

```python
table=pd.DataFrame.from_dict(re)
table
```
